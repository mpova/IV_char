```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

# RED DIODE

```python
df = pd.read_csv('reddiode.csv', delimiter=';')
df.info()
```

```python
df
```

```python
df = df.sort_values(by='V_in (Volt)')
```

```python
df.plot(y='I (milliamp)', x = 'V')
plt.xlabel('I (milliamp)')
plt.ylabel('V')
plt.title('I-V Plot')

# Show the plot
plt.grid(True)
plt.show()
```
